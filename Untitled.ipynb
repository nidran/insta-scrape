```python
import json
import bs4
import requests
import re
import logging as log
from abc import ABCMeta, abstractmethod
import sys
```

```python
url_string = "https://www.instagram.com/explore/tags/food/"
```

```python
response = bs4.BeautifulSoup(requests.get(url_string).text, "html.parser")
```

```python
response
```

```python
def get_query_id( doc):
        query_ids = []
        for script in doc.find_all("script"):
            if script.has_attr("src") and "en_US_Commons" in script['src']:
                text = requests.get("%s" % ( script['src'])).text
                for query_id in re.findall("(?<=queryId:\")[0-9]{17,17}", text):
                    query_ids.append(query_id)
        return query_ids
```

```python
potential_query_ids = get_query_id(response)
```
